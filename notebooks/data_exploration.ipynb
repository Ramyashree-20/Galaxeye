```python
import os
import cv2
import numpy as np

train_path = "../data/train/train"

pre_path = os.path.join(train_path, "pre-event")
post_path = os.path.join(train_path, "post-event")
mask_path = os.path.join(train_path, "target")

sample_name = os.listdir(pre_path)[0]

print("Sample file:", sample_name)

pre_image = cv2.imread(os.path.join(pre_path, sample_name))
post_image = cv2.imread(os.path.join(post_path, sample_name))
mask = cv2.imread(os.path.join(mask_path, sample_name), 0)

print("Pre-image shape:", pre_image.shape)
print("Post-image shape:", post_image.shape)
print("Mask shape:", mask.shape)

print("Unique mask values:", np.unique(mask))
```
